```python
import cv2
photo = cv2.imread('/content/Sidharth-Malhotra-EE.jpg')
```

```python
import time
start = time.time()
```

```python
gray = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)
from google.colab.patches import cv2_imshow
cv2_imshow(gray)
```

```python
invert = cv2.bitwise_not(gray)
cv2_imshow(invert)
```

```python
blur = cv2.GaussianBlur(invert, (15, 15), 0)
cv2_imshow(blur)
```

```python
invertblur = cv2.bitwise_not(blur)
cv2_imshow(invertblur)
```

```python
sketch = cv2.divide(gray, invertblur, scale= 250.0)
cv2_imshow(sketch)
```
